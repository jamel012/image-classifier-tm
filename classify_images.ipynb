### **test kiwi file**

```python
from keras.models import load_model
from PIL import Image, ImageOps
import numpy as np
import os
from collections import defaultdict

# Disable scientific notation for clarity
np.set_printoptions(suppress=True)

# Load the model
model = load_model("/content/drive/MyDrive/keras_model.h5", compile=False)

# Load the labels
class_names = [label.strip() for label in open("labels.txt", "r").readlines()]

# Set the image folder path
image_folder = "/content/drive/MyDrive/test/kiwi"
size = (224, 224)

# Counters
total_images = 0
class_counts = defaultdict(int)
class_confidence_sum = defaultdict(float)

# Loop through each image in the folder
for filename in os.listdir(image_folder):
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):
        total_images += 1
        image_path = os.path.join(image_folder, filename)

        # Load and preprocess the image
        image = Image.open(image_path).convert("RGB")
        image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)
        image_array = np.asarray(image)
        normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1
        data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)
        data[0] = normalized_image_array

        # Predict
        prediction = model.predict(data)
        index = np.argmax(prediction)
        class_name = class_names[index]
        confidence_score = prediction[0][index]

        # Update stats
        class_counts[class_name] += 1
        class_confidence_sum[class_name] += confidence_score

        # Print result
        print(f"Image: {filename}")
        print(" → Class:", class_name)
        print(" → Confidence Score:", round(confidence_score * 100, 2), "%")
        print("-" * 40)

# Final summary
print("\nSummary:")
print("Total images processed:", total_images,"\nactual feature is kiwi")
print("\nClass breakdown:")
for class_name in class_names:
    if class_name == "1 watermelon":
        print("total wrong choice :-")
    count = class_counts[class_name]
    if count > 0:
        avg_conf = (class_confidence_sum[class_name] / count) * 100
        print(f"- {class_name}: {count} images, Avg Confidence: {round(avg_conf, 2)}%")

    else:
        print(f"- {class_name}: 0 images")
```
